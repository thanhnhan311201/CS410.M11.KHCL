## Khởi tạo API

```python
api = create(f'{root}/NAS_Benchmark/datasets/NATS-tss-v1_0-3ffb9-simple', 'tss', fast_mode=True)
api.verbose = False
```

## Ghi chép toàn bộ kiến trúc

```python
def record_entire_arch( dataset, hp ):
  with open(f'{root}/NAS_Benchmark/datasets/NAS_201_{dataset}_{hp}.txt', 'a') as fi:
    for i in range(11074, 15625):
      arch_info = dict()
      info = api.get_more_info(i, dataset=dataset, hp=str(hp), is_random=False)
      cost = api.get_cost_info(i, dataset=dataset, hp=str(hp))
      arch_info['arch_index'] = i
      arch_info['arch'] = api.arch(i)
      arch_info['test_accuracy'] = np.round(info['test-accuracy'], 2)
      arch_info['parameters'] = cost['params']
      string = str(arch_info).replace("'", '"')
      print(string)
      fi.write(string + '\n')

  return 'Done!'
```

```python
datasets = ['cifar100']
check_points = [200]
for dataset in datasets:
  for check_point in check_points:
    record_entire_arch(dataset, check_point)
```
